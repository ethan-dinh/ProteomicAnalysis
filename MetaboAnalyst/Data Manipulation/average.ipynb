```python
# Importing Libraries
import pandas as pd

# Declaring Data Path
dirPath = "../../Input Data/Split Time Points"

# Initializing a diction of split times
timeSeries = {}
for visit in range(1, 6):
    timeSeries[visit] = pd.read_csv(f"{dirPath}/Soma plus MGH repeat measures (T{visit}).csv")
```

```python
# Initializing the output file
preREFX = timeSeries[1].copy()
postREFX = timeSeries[4].copy()

# Altering time signature on sampleID
for index in range(len(preREFX.index)):
    preREFX.iloc[index, 0] = f"{preREFX['SampleId'][index][:-2]} (T1-T3)"

for index in range(len(postREFX.index)):
    postREFX.iloc[index, 0] = f"{postREFX['SampleId'][index][:-2]} (T4-T5)"

# Displaying modified output files
display(preREFX.head(1))
display(postREFX.head(1))
```

```python
# Retrieving the column names
col_names = preREFX.columns.tolist()
col_names.remove("visit")
col_names = col_names[1:1513]
```

```python
# Averaging by indexing through the time points – Pre-refracture
for feat in col_names:
    for row_index in range(len(preREFX.index)):    
        col_index = preREFX.columns.get_loc(feat)
        average = (timeSeries[1].iloc[row_index, col_index] + timeSeries[2].iloc[row_index, col_index] + timeSeries[3].iloc[row_index, col_index]) / 3
        preREFX.iloc[row_index, col_index] = average
```

```python
# Averaging by indexing through the time points – Pre-refracture
for feat in col_names:
    for row_index in range(len(postREFX.index)):    
        col_index = postREFX.columns.get_loc(feat)
        average = (timeSeries[4].iloc[row_index, col_index] + timeSeries[5].iloc[row_index, col_index]) / 2
        postREFX.iloc[row_index, col_index] = average
```

```python
# Displaying modified output files
display(preREFX.head(1))
display(postREFX.head(1))

# Saving to directory
outPath = "../../Input Data/Averaged Time Points (Pre:Post Refracture)"
preREFX.to_csv(f"{outPath}/Soma plus MGH repeat measures (AVG T1-T3).csv", index=False)
postREFX.to_csv(f"{outPath}/Soma plus MGH repeat measures (AVG T4-T5).csv", index=False)
```
